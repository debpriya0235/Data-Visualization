```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
data.dropna(inplace=True) # drop the rows with missing values
data.isnull().sum() # check for missing values again
```

```python
# i) Check for missing values
print("Missing values in each column:")
print(data.isnull().sum())

# ii) Replace missing numerical values
for column in data.select_dtypes(include=['float64', 'int64']).columns:
    if data[column].isnull().any():
        # Replace with mean
        data[column].fillna(data[column].mean(), inplace=True)
        # Or replace with median
        # data[column].fillna(data[column].median(), inplace=True)

print("\nData after replacing missing values:")
print(data)

```

```python
# i) Pairplot
sns.pairplot(data)
plt.suptitle('Pairplot of World Cup Data', y=1.02)
plt.show()

```

```python
# ii) Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Year', y='Attendance', data=data)  # Changed 'attendence' to 'Attendance'
plt.title('Boxplot of Attendance by Year')
plt.xticks(rotation=90)
plt.show()
```

```python
# iii) Heatmap
plt.figure(figsize=(10, 6))
# Select only numeric columns for correlation calculation
numeric_data = data.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of World Cup Data Correlations')
plt.show()
```

```python
# iv) Barplot
plt.figure(figsize=(10, 6))
# Verify the correct column name for 'match id' in your DataFrame
# and replace 'MatchID' below with the correct name
sns.barplot(x='Year', y='MatchID', data=data)
plt.title('Barplot of match id')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(x='Home Team Goals',data=data) #show histplot
```

```python
# v) Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data['Attendance'], bins=10, kde=True) # Changed 'attendence' to 'Attendance'
plt.title('Histogram of Attendance') # Changed title to reflect correct column name
plt.show()
```

```python
# vii) Scatterplot
sns.scatterplot(x='Home Team Goals', y='Away Team Goals', data=data,label='Home_Team')
plt.title('Scatterplot of Home Team Goals vs. Away Team Goals')
plt.show()
```
